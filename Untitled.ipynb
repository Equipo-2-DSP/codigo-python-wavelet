# Encriptado de Imagenes BMP con Wavelet

```python
import pywt
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
def rgb2gray(rgb):
    return np.round( np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140]) * 255 )

def plot_img( img ):
    plt.imshow( img, cmap=plt.cm.gray )
```

```python
img = mpimg.imread( './Inputs/monte.jpeg' )
img = rgb2gray( img )
plot_img( img )
```

```python
img_transform = pywt.wavedec2( img, 'haar', level=1 )
```

```python
LL, ( HL, LL, HH ) = img_transform
```

```python
fig, axs = plt.subplots( 1, 4, figsize=(40,10) )
for i, img_ in enumerate([ LL, HL, LL, HH ]):
    axs[i].imshow( img_, cmap=plt.cm.gray )
```

```python
rows, cols = LL.shape
r_ll = np.random.randint( 0, 255, ( rows, cols ) )

masc = np.real( np.exp( 2 * np.pi * 1j * r_ll ) )
LL_ = LL * masc
```

```python
plot_img( LL_ )
```
