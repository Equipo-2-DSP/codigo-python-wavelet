```python
import matplotlib.pyplot as plt
from filterw import ImageFilter, thresholds
import cv2
import numpy as np
plt.rcParams.update({
    'figure.figsize': (10, 10)
})
```

```python
img = cv2.imread( 'Inputs/face_01.jpg', cv2.IMREAD_GRAYSCALE ) / 255
def noise( img, std ):
    row, col = img.shape
    mean = 0
    gauss = np.random.normal( mean, std ** 2, (row, col) )
    gauss = gauss.reshape( row, col )
    img_noise = img + gauss 
    return img_noise / np.max( img_noise )
img = noise( img, .3 )
imageFilter = ImageFilter.ImageFilter( thresholds.universal_threshold, 'unique', img )
```

```python
img_desnoise = imageFilter.filter( wavelet='sym4' )
```

```python
plt.imshow( img, cmap=plt.cm.gray )
```

```python
plt.imshow( img_desnoise, cmap=plt.cm.gray )
```
