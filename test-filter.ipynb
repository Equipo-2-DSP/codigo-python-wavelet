```python
import matplotlib.pyplot as plt
from filterw import ImageFilter, thresholds, noise
import cv2
import numpy as np
plt.rcParams.update({
    'figure.figsize': (10, 10)
})
```

```python
img = cv2.imread( 'Inputs/face_01.jpg', cv2.IMREAD_GRAYSCALE ) / 255

img = noise.noise_2d( img, .3 )
imageFilter = ImageFilter.ImageFilter( thresholds.bayes_threshold, 'for-levels', img )
imageFilter.threshold_types()
```

```python
img_desnoise = imageFilter.filter( wavelet='sym4' )
```

```python
plt.imshow( img, cmap=plt.cm.gray, vmin=0, vmax=1 )
```

```python
plt.imshow( img_desnoise, cmap=plt.cm.gray, vmin=0, vmax=1 )
```

```python
img_desnoise += np.abs( np.min( img_desnoise ) )
img_desnoise /= np.max( np.max( img_desnoise ) )
cv2.imwrite( 'blade_desnoise.png', (img_desnoise*255).astype( np.uint8 ) )
```

```python
img_desnoise = img
img_desnoise += np.abs( np.min( img_desnoise ) )
img_desnoise /= np.max( np.max( img_desnoise ) )
cv2.imwrite( 'blade_noise.png', (img_desnoise*255).astype( np.uint8 ) )
```
