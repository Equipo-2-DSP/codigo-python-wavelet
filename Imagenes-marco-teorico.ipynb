```python
import numpy as np 
import matplotlib.pyplot as plt
import pywt
```

```python
x = lambda t: np.sin( 2 * np.pi * t ) + np.sin( 2 * np.pi * 2 * t )
doppler = pywt.data.demo_signal('doppler', 1024)
```

```python
t = np.linspace( 0, 4, 3000 )
```

```python
fig, axs = plt.subplots( 1, 2, figsize=(10, 5) )
axs[0].plot( t, x(t) )
axs[1].plot( doppler )

axs[0].set_xticks([])
axs[1].set_xticks([])

axs[0].set_yticks([])
axs[1].set_yticks([])

axs[0].grid()
axs[1].grid()

axs[0].set_title( 'Señal estacionaria', fontsize=20 )
axs[1].set_title( 'Señal no estacionaria', fontsize=20 )

axs[0].set_ylabel( 'Amplitud', fontsize=20 )
axs[1].set_ylabel( 'Amplitud', fontsize=20 )

axs[0].set_xlabel( 't', fontsize=20 )
axs[1].set_xlabel( 't', fontsize=20 )

plt.savefig('../informe/Img/estacionaria-no-estacionaria.png')
```
